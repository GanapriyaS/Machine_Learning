```python
heart_disease= pd.read_csv("../datasets/heart-disease.csv")
heart_disease.head()
```

```python
over_50=heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
#Pyplot method
over_50.plot(kind='scatter',x='age',y='chol',c='target',figsize=(10,6));
```

```python
#OO method mixed from pyplot method
fix,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",x='age',y='chol',c='target',ax=ax);

ax.set_xlim([45,100]);
```

```python
over_50.target.values
```

```python
#OO method mixed from scratch method
fix,ax=plt.subplots(figsize=(10,6))

#plot data
scatter=ax.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target']);

#Customise the plot
ax.set(title="Heart Disease and cholesterol level",xlabel="Age",ylabel="Cholesteral");

#Add legend
ax.legend(*scatter.legend_elements(),title="Target");

#Add a horizontal line
ax.axhline(over_50["chol"].mean(),linestyle="--");
```

```python
fig,(ax0,ax1) = plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)

#add data to ax0
scatter = ax0.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target']);

#Customise the plot ax0
ax0.set(title="Heart Disease and cholesterol level",xlabel="Age",ylabel="Cholesteral");

#Add legend
ax0.legend(*scatter.legend_elements(),title="Target");

#Add a mean line
ax0.axhline(over_50["chol"].mean(),linestyle="--");

#add data to ax1
scatter = ax1.scatter(x=over_50['age'],y=over_50['thalach'],c=over_50['target']);

#Customise the plot ax1
ax1.set(title="Heart Disease and Max heart rate",xlabel="Age",ylabel="Max heart rate");

#Add legend
ax1.legend(*scatter.legend_elements(),title="Target");

#Add a mean line
ax1.axhline(over_50["thalach"].mean(),linestyle="--");

#Add title to figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```

```python
#set the style
plt.style.use('seaborn-whitegrid')

#OO method mixed from scratch method
fix,ax=plt.subplots(figsize=(10,6))

#plot data
scatter=ax.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'],cmap="winter");

#Customise the plot
ax.set(title="Heart Disease and cholesterol level",xlabel="Age",ylabel="Cholesteral");

#Add legend
ax.legend(*scatter.legend_elements(),title="Target");

#Add a horizontal line
ax.axhline(over_50["chol"].mean(),linestyle="--");
```

```python
#customising y and x axis limitations
fig,(ax0,ax1) = plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)

#add data to ax0
scatter = ax0.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'],cmap="summer");

#Customise the plot ax0
ax0.set(title="Heart Disease and cholesterol level",xlabel="Age",ylabel="Cholesteral");

#change the xaxis limits
ax0.set_xlim([50,80])

#Add legend
ax0.legend(*scatter.legend_elements(),title="Target");

#Add a mean line
ax0.axhline(over_50["chol"].mean(),linestyle="--");

#add data to ax1
scatter = ax1.scatter(x=over_50['age'],y=over_50['thalach'],c=over_50['target'],cmap="plasma");

#Customise the plot ax1
ax1.set(title="Heart Disease and Max heart rate",xlabel="Age",ylabel="Max heart rate");


#change the xaxis limits
ax1.set_xlim([50,80]) #not necessary since a0 set and since both plots share an xaxis
ax1.set_ylim([80,200])
             
#Add legend
ax1.legend(*scatter.legend_elements(),title="Target");

#Add a mean line
ax1.axhline(over_50["thalach"].mean(),linestyle="--",color="violet");

#Add title to figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```
