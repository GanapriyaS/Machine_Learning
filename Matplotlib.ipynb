```python
fig,ax= plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
#Plot from dictionary (bar plot)
a={'Almond':10, 'peanut':8, 'cashew':9}
fig,ax= plt.subplots()
ax.bar(a.keys(),a.values())
# ax.bar(a.keys(),height=a.values());
ax.set(title="NUTS",ylabel="Price ($)");
```

```python
fig,ax= plt.subplots()
# list is important
ax.barh(list(a.keys()),list(a.values()));
```

```python
# normal distribution (histogram)
x=np.random.randn(1000)
fig,ax=plt.subplots()
ax.hist(x);
```

## Subplots

```python
# option 1
# 4 in one commmand it is flexibility of object oriented API
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
#Plot to differnt axis
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(a.keys(),a.values());
ax4.hist(np.random.randn(1000));
```

```python
#option2
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
#Plot to differnt axis
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(a.keys(),a.values());
ax[1,1].hist(np.random.randn(1000));
```

## Pandas DataFrame

```python
df=pd.read_csv("Datasets/car-sales.csv")         
df
```

```python
df.dropna(inplace=True)
df
```

```python
ts=pd.Series(np.random.randn(1000),index=pd.date_range("1/1/2020",periods=1000))
ts
```

```python
ts=ts.cumsum()
#return cumulative sum over a dataframe or series axis
ts.plot();
```

```python
df['Price']=df["Price"].str.replace('[\$\,\.]',"",regex=True)
df
```

```python
type(df['Price']),type(df['Price'][0])
```

```python
df["Price"]=df["Price"].str[:-2]
#remove last 2 zeros
df
```

```python
df["Sale Date"]=pd.date_range("1/1/2020",periods=len(df))
df
```

```python
df["Total Sales"]=df["Price"].astype(int).cumsum()
df
```

```python
#pandas version of matplotlib API
df.plot(x="Sale Date",y="Total Sales");
```

```python
df["Price"]=df["Price"].astype(int)
df.plot(x="Odometer (KM)",y="Price",kind="scatter");
```
