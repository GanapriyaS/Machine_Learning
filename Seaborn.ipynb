```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
# setting the standard color code for seaborn package
%matplotlib inline
#remove the use of plt.show() everytime
```

```python
auto=pd.read_csv('Datasets/Automobile.csv')
# include datasets
auto.dropna(inplace=True)
# remove null values
```

```python
auto
```

```python
auto.head()
# look the head of data
```

# Univariate (Single variable)
Looking at specific variables within the dataset and analyzing then at a univariate structure

# Distance plot in seaborn
Histogram

```python
sns.displot(auto['normalized_losses'])
# distplot() or histplot() can be used
plt.show()
# kernal density present
```

```python
sns.histplot(auto['symboling'])
#displot()
plt.show()
```

```python
sns.displot(auto['normalized_losses'],kde=False,rug=True)
plt.show()
# rug if more denser,more dataset there
```

# Bivariate Distribution
Between 2 continuous variables

```python
sns.jointplot(x=auto['engine_size'],y=auto['horsepower'])
#linear relationship  for engine-size and horse 
plt.show()
# positive correlation R value about 0.82
```

```python
sns.jointplot(x=auto['engine_size'],y=auto['horsepower'],kind="hex")
plt.show()
# kind==={"scatter"| "reg"(regression) |"resid"(residual)| "kde"| "hex"(hexagon)}
# concentration of hexagons darker-- higher concentration
```

```python
sns.jointplot(x=auto['engine_size'],y=auto['horsepower'],kind="resid")
plt.show()
# residual plot-- plotting error or residuals between engine size and horsepower linear regression eqs
```

```python
sns.jointplot(x=auto['engine_size'],y=auto['horsepower'],kind="reg")
plt.show()
# this line is used to create residual plot
# residual calculated by distance bw best fit line and actual observations
```

```python
sns.jointplot(x=auto['engine_size'],y=auto['horsepower'],kind="kde")
plt.show()
# densities are ploted
```

# Multivariate distribution
Between multiple continuous variables

```python
sns.pairplot(auto[['normalized_losses','engine_size','horsepower']])
#relationship between all given in ()
# no need of show() as matplotlib inline declared
# upper and lower triangle are mirrror image of eachother so one is enough to watch
```

# 2 D plot categorical and continous type

```python
sns.stripplot(x=auto['fuel_type'],y=auto['horsepower'],jitter=False)
#ordered scatter plot as categorical variables in xaxis instead of continuous variables
```

```python
sns.stripplot(x=auto['fuel_type'],y=auto['horsepower'],jitter=True)
# jitter show the concentration of plot
# i.e one car has high horsepower in gas data
```

```python
sns.swarmplot(x=auto['fuel_type'],y=auto['horsepower'])
# no overlap over the data
#show the frequency of data in each level
```

```python
sns.boxplot(x=auto['number_of_doors'],y=auto['horsepower'])
#certain data and anomaly cannot be detected using boxplot
# Q1 25 ,Q2 50 ,Q3 75 ,i.e 1.5 times the data ,but 
# 100 percent not follow rule of 1.5 times interquartile range dist bw Q3 and Q1 and distance between 75to observation(more than 1.5)
# i.e automatically converted into outliers by box plot logic
```

```python
sns.boxplot(x=auto['number_of_doors'],y=auto['horsepower'],hue=auto['fuel_type'])
# extra information with categorical variable
```

```python
# bar graph use concept of bootstrap or aggregation to compute confidence interval around the estimated valuse and plot using error bars
# There is one error bar shown at the top tell confidence intervals
# include 0 in quantitive range
# lines are confident interval for distribution interval
sns.barplot(x=auto['body_style'],y=auto['horsepower'])
```

```python
sns.barplot(x=auto['body_style'],y=auto['horsepower'],hue=auto['engine_location'])
```

```python
# within a bar plot special case ,instead of show of cont variabel on Yaxis, want to see frequency of observations for a set of categorical variable or for asingle categorical variable
# show the frequncy of observation in terms of hatchback in terms of sedan,wagon and hardtop
sns.countplot(x=auto['body_style'])
```

```python
sns.countplot(x=auto['body_style'],hue=auto['engine_location'])
```

```python
# bar plot from 0 to highest point, point plot show only highest point
# show confidence interval if applicable for following data
# show line trend as connect the data
sns.pointplot(x=auto['fuel_system'],y=auto['horsepower'])
```

```python
sns.pointplot(x=auto['fuel_system'],y=auto['horsepower'],hue=auto['number_of_doors'])
```

```python
sns.pointplot(auto['fuel_system'],auto['horsepower'],hue=auto["number_of_cylinders"])
```

```python
# factor plot used to draw multipanel categorical plot
sns.factorplot(x="fuel_type",
                  y="horsepower",
                  hue="number_of_doors",
                  col="engine_location",
                  data=auto,
                  kind="swarm")
```

```python
sns.factorplot(x="fuel_type",
                  y="horsepower",
                  hue="number_of_doors",
                  col="engine_location",
                  data=auto,
                  kind="point")
```

```python
# factorplot is as same as carplot
sns.catplot(x="fuel_type",
                  y="horsepower",
                  hue="number_of_doors",
                  col="engine_location",
                  data=auto,
                  kind="bar")
```

```python
# count is only y or x
sns.factorplot(x="fuel_type",
                  hue="number_of_doors",
                  col="engine_location",
                  data=auto,
                  kind="count")
```

```python
sns.catplot(   y="horsepower",
                  hue="number_of_doors",
                  col="engine_location",
                  data=auto,
                  kind="count")
```

```python
sns.catplot(x="fuel_type",
                  y="horsepower",
                  hue="number_of_doors",
                  col="engine_location",
                  data=auto,
                  kind="box")
```

```python
# violin plot is as same as box plot creating kernal density around box plot
sns.catplot(x="fuel_type",
                  y="horsepower",
                  hue="number_of_doors",
                  col="engine_location",
                  data=auto,
                  kind="violin")
sns.catplot(x="fuel_type",
                  y="horsepower",
                  hue="number_of_doors",
                  col="engine_location",
                  data=auto,
                  kind="strip")
```

# Linear Regression



```python
sns.lmplot(x="horsepower", y="peak_rpm", data=auto)
# correlation weak as slight horizontal and confidence interval expanding bw 2 values
```

```python
sns.lmplot(x="horsepower", y="peak_rpm", data=auto,hue="fuel_type")
# deisel has negative coorelation bw horsepower and peak rpm
#As horse power increses peak rpm decreases in deisel
# in gas somewhat neutral correlation
```
